```python
import pandas as pd
import plotly.express as px
import plotly

DATA_URL = "https://raw.githubusercontent.com/ArushiAg22/arushi_data690/main/Individual_Assignment_Part_B/March_28_690.csv"
```

```python
df  = pd.read_csv(DATA_URL)

df.sample(5)
```

```python
df_2020_pop = df.query("Year == 2020").query("indicator == 'SP.POP.TOTL'")
df_2020_pop.sample(5)
```

```python
df_2020_pop_sort = df_2020_pop.sort_values(by= 'value', ascending=False)
```

```python
fig = px.bar(
    data_frame= df_2020_pop_sort,
    x= "Country Name",
    labels={"value":"2020 Population", "Country Name":""},
    y= "value",
    color= "Country Name",
    height= 800,
    template=list(plotly.io.templates.keys())[5]              ## template='plotly_dark'
)

fig.update_layout(showlegend= False)
fig.show()
```

```python
df_group_sum = df_2020_pop.groupby("Region").sum().reset_index()

df_group_sum
```

```python
df_group_sum_sorted = df_group_sum.sort_values(by= "value", ascending= False)
```

```python
fig = px.bar(
    data_frame= df_group_sum_sorted,
    x= "Region",
    labels={"value":"2020 Population", "Region":""},
    y= "value",
    color= "Region",
    height= 800,
    template=list(plotly.io.templates.keys())[5],
    title= "Comparing Total Populations on All Continents"
)

fig.update_layout(showlegend= False)
fig.show()
```

```python
df_gdp = df.query("indicator == 'NY.GDP.PCAP.PP.CD'")

df_gdp = df_gdp[df_gdp["Country Code"].isin(["CHN", "IND", "RUS", "USA"])]
df_gdp["Year"]= df_gdp["Year"].astype('str')

df_gdp
```

```python
df_gdp.info()
```

```python
df_gdp2 = df_gdp
df_gdp2["Year"] = df_gdp2["Year"].astype("int")

fig = px.line(
    df_gdp2,
    x= "Year",
    labels={"value":"GDP Per Capita"},
    y= "value",
    color= "Country Name",
    height= 800,
    template=list(plotly.io.templates.keys())[5],
    title= "Comparing GDP Per Capita from 2016 to 2020 of Four Major Economies"
)

for index, row in df_gdp2.query("Year == '2020'").iterrows():
  anno_dict = dict(
      font= dict(color= "white"),
      x= row["Year"],
      y= row["value"],
      showarrow= False,
      text= row["Country Name"],
      textangle= 0,
      xanchor= 'left',
      yanchor= 'bottom',
      xref= 'x',
      yref= 'y'
  )

fig.add_annotation(anno_dict)

fig.show()
```
