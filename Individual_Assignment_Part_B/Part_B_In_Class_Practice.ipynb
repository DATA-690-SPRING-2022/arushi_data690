```python
import pandas as pd
import plotly.express as px
import plotly



DATA_URL = "https://raw.githubusercontent.com/ArushiAg22/arushi_data690/main/Individual_Assignment_Part_B/March_28_690.csv"
```

```python
df  = pd.read_csv(DATA_URL)

df.sample(5)
```

```python
df_2020_pop = df.query("Year == 2020").query("indicator == 'SP.POP.TOTL'")
df_2020_pop.sample(5)
```

```python
df_2020_pop_sort = df_2020_pop.sort_values(by= 'value', ascending=False)
```

```python
fig = px.bar(
    data_frame= df_2020_pop_sort,
    x= "Country Name",
    labels={"value":"2020 Population", "Country Name":""},
    y= "value",
    color= "Country Name",
    height= 800,
    template=list(plotly.io.templates.keys())[5]              ## template='plotly_dark'
)

fig.update_layout(showlegend= False)
fig.show()
```

```python
df_group_sum = df_2020_pop.groupby("Region").sum().reset_index()

df_group_sum
```

```python
df_group_sum_sorted = df_group_sum.sort_values(by= "value", ascending= False)
```

```python
fig = px.bar(
    data_frame= df_group_sum_sorted,
    x= "Region",
    labels={"value":"2020 Population", "Region":""},
    y= "value",
    color= "Region",
    height= 800,
    template=list(plotly.io.templates.keys())[5],
    title= "Comparing Total Populations on All Continents"
)

fig.update_layout(showlegend= False)
fig.show()
```

```python
df_gdp = df.query("indicator == 'NY.GDP.PCAP.PP.CD'")

df_gdp = df_gdp[df_gdp["Country Code"].isin(["CHN", "IND", "RUS", "USA"])]

df_gdp
```

```python
df_gdp.info()
```

```python
fig = px.line(
    df_gdp,
    x= "Year",
    labels={"value":"GDP Per Capita"},
    y= "value",
    color= "Country Name",
    height= 800,
    template=list(plotly.io.templates.keys())[5],
    title= "Comparing GDP Per Capita from 2016 to 2020 of Four Major Economies"
)

fig.update_xaxes(dtick="Y1")
fig.show()
```

```python
fig = px.line(
    df_gdp, 
    x="Year", 
    labels={"value":"GDP Per Capita"},
    y="value",
    color="Country Name",
    height=800,
    template=list(plotly.io.templates.keys())[5],
    title="Comparing GDP Per Capita from 2016 to 2020 of Four Major Economies"
)


_df = df_gdp.query("Year == 2020")
for index, row in _df.iterrows():
    anno_dict = dict(
        font=dict(color="white"),
        x=row["Year"],
        y=row["value"],
        showarrow=False,
        text=row["Country Name"],
        textangle=0,
        xanchor='left',
        yanchor="bottom",
        xref="x",
        yref="y"
    )
    fig.add_annotation(anno_dict)  

fig.update_xaxes(dtick="Y1")       # tick interval very year.
fig.update_layout(showlegend=False)
fig.show()
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)

df.head()
```

```python
df_pivot = df.pivot_table(
    index=["Year", "Country Code",	"Country Name",	"Region",	"Income Group",	"Lending Type"],
    columns="indicator",
    values="value")

df_pivot
```

```python
df_pivot.reset_index(inplace=True)

df_pivot.head()
```

```python
df_scatter = df_pivot[df_pivot["Country Code"].isin(["CHN", "IND", "RUS", "USA"])].query("Year == 2018")

fig = px.scatter(
    df_scatter,
    x= "NY.GDP.PCAP.PP.CD",
    y= "SP.DYN.LE00.IN",
    size="SP.POP.TOTL",
    color= "Region",
    labels={"NY.GDP.PCAP.PP.CD": "GDP Per Capita", "SP.DYN.LE00.IN": "Life Expectancy"},
    text= "Country Name"
    )

fig.update_traces(textposition= "top left")
fig.show()
```
