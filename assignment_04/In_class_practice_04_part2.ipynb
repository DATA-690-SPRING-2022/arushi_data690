```python
def cleanse_year(year_str):
    if "*" in year_str:
        year = year_str.replace("*", "")
    else:
        year = year_str

    return int(year)
```

```python
cleanse_year("2010*")
```

```python
type(cleanse_year("2010*"))
```

```python
print(cleanse_year("2010*"))
print(type(cleanse_year("2010*")))
```

```python
print(cleanse_year("2000"))
print(type(cleanse_year("2000")))
```

```python
df["year_cleansed"] = df["Census Year"].apply(cleanse_year)

df
```

```python
def cleanse_pop(pop_str):
    if "," in pop_str:
        pop = pop_str.replace(",", "")
    else:
        pop = pop_str

    return int(pop)
```

```python
cleanse_pop("3,929,214")
```

```python
df["population_cleansed"] = df["Total Population"].apply(cleanse_pop)

df
```
